```python
# starting with unsupervised algorithm : Isolation Forest

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# add new features : sec, minute, hour, day
df["sec"]= df.Time%60
df["minute"]=(df.Time%3600)//60
df["hour"]=(df.Time%(24*3600))//3600
df["day"]= df.Time//(24*3600)
```

```python
df.columns
```

```python
df.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])

plt.show()
```

```python
colors = ["#0101DF", "#DF0101"]

sns.countplot('Class', data=df, palette=colors)
plt.title('Class Distributions \n (0: No Fraud || 1: Fraud)', fontsize=14)
```

```python
sns.catplot(data=df, x="Class", y="Amount", kind="box")
```

```python
median_nofraud = df[df.Class==0].median()
median_fraud = df[df.Class==1].median()
```

```python
median_fraud['Time']
```

```python
cols = [e for e in median_nofraud.index if abs(median_fraud[e]-median_nofraud[e]) > 4 ]
```

```python
# data preparation
labels = df.Class
#data = df.drop(columns=['Time', 'Class'])
data = df[cols]
data = data.drop(columns=['Time', 'Amount'])
```

```python
data.columns
```

```python
# scaling data
scaler = StandardScaler()
scaler.fit(data)
X = scaler.transform(data) 
```

```python
# training IF model
from sklearn.ensemble import IsolationForest 

clf = IsolationForest(random_state=64, contamination=0.1)
clf.fit(X)
y_pred = clf.predict(X)

y_pred[y_pred==1] = 0
y_pred[y_pred==-1] = 1
```

```python
# plotting results
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score, recall_score, precision_score

cm = confusion_matrix(labels, y_pred, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["normal", "anomalous"])
disp.plot()
```

```python
# Analyse performances
F1 = f1_score(labels, y_pred)
Recall = recall_score(labels, y_pred)
Precision = precision_score(labels, y_pred)

print('# F1 score = %.2f'%F1)
print('# Recall score = %.2f'%Recall)
print('# Precision score = %.2f'%Precision)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
RFC_METRIC = 'gini'  #metric used for RandomForrestClassifier
NUM_ESTIMATORS = 10 #number of estimators used for RandomForrestClassifier
NO_JOBS = 4 #number of parallel jobs used for RandomForrestClassifier
RANDOM_STATE = 2018
```

```python
clf = RandomForestClassifier(n_jobs=NO_JOBS, 
                             random_state=RANDOM_STATE,
                             criterion=RFC_METRIC,
                             n_estimators=NUM_ESTIMATORS,
                             verbose=False)
clf.fit(X, labels)
```

```python
y_pred = clf.predict(X)
```

```python
# plotting results
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score, recall_score, precision_score

cm = confusion_matrix(labels, y_pred, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["normal", "anomalous"])
disp.plot()
```

```python
# Analyse performances
F1 = f1_score(labels, y_pred)
Recall = recall_score(labels, y_pred)
Precision = precision_score(labels, y_pred)

print('# F1 score = %.2f'%F1)
print('# Recall score = %.2f'%Recall)
print('# Precision score = %.2f'%Precision)
```
